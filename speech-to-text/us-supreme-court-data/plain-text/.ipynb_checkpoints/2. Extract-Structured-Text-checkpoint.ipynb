```python
i = 0
for i in range(0,len(all_text),5):
    print(all_text[i:i+5])
    result = text_analytics_client.recognize_entities(all_text[i:i+5])
    result = [court_sent for court_sent in result if not court_sent.is_error]
    all_result.append(result)
 
```

```python
entities_with_values = dict()
```

```python
for cat_entities in all_result:
    print(cat_entities)
    break
```

```python
for entry in all_result:
    for cat_entities in entry:
        if len(cat_entities.entities) != 0:
            if cat_entities.entities[0].category in entities_with_values:
                entities_with_values[cat_entities.entities[0].category].append(cat_entities.entities[0].text)
            else:
                entities_with_values[cat_entities.entities[0].category] = [cat_entities.entities[0].text]
```

```python
def strip_non_ascii(string):
    stripped = (c for c in string if 0 < ord(c) < 127)
    return ''.join(stripped)
```

```python
with open("../structured-text/us_supreme_structured_text.md", "w", encoding="utf_8_sig") as f:
    for k in entities_with_values.keys():
        f.write("@ list " + k + " = ")
        f.write("\n")
        list_of_entities = list(set(entities_with_values[k]))
        for entity in list_of_entities:
            if len(entity) >= 2:
                entity = entity.strip()
                entity = strip_non_ascii(entity)
                f.write("- " + entity.strip())
                f.write("\n")
        
```

## Read from Structured Text File and load all the lines into a Python object

```python
# read entities from structured text file and apply to sentences 
with open("../structured-text/us_supreme_structured_text.md", "r", encoding="utf_8_sig") as f:
    lines = f.readlines()
```

```python
with open("languagemodeltraintext.txt", "r", encoding="utf_8_sig") as f:
    all_sentences = f.readlines()
```

```python
entity_dict = {}
key = ""
```

```python
for line in lines:
    if line[0] == '@':
        key = line.split(' ')[2]
        entity_dict[key] = []
    elif line[0] == '\n':
        print("Added entity: " + key + " to dictionary")
    else:
        entity_val = ' '.join(line.split(' ')[1:len(line.split(' '))]).rstrip()
        entity_val = strip_non_ascii(entity_val)
        if len(entity_val) >= 2:
            entity_dict[key] += [entity_val]
        
```

```python
entity_dict
```

```python
replaced_sentences = [] 
```

```python
for sentence_line in all_sentences:
    sentence_line = sentence_line.rstrip()
    for key, val in entity_dict.items():
        to_replace_with = "{@" + key + "}"
        for entity in val:
            to_replace = entity
            is_match = re.search(to_replace,sentence_line)
            if is_match != None:
                sentence_line = re.sub(to_replace,to_replace_with,sentence_line) 
    replaced_sentences.append(sentence_line)
```

```python
replaced_sentences
```

```python
with open("../structured-text/us_supreme_structured_text.md", "a", encoding="utf_8_sig") as f:
    f.write("\n")
    f.write("\n")
    f.write("# SomeTrainingSentence")
    f.write("\n")
    for sentence in replaced_sentences:
        if len(sentence) > 2:
            sentence = strip_non_ascii(sentence)
            f.write("- " + sentence.strip())
            f.write("\n")
```
